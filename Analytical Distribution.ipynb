```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import thinkstats2
import thinkplot

df = pd.read_csv('bad-drivers.csv')

fatal_billion = df['Number of drivers involved in fatal collisions per billion miles']
fatal_speed = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding']
fatal_alcoholimpared = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired']
fatal_notdistracted = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted']
fatal_notprevaccidents = df['Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents']
```

This is a Normal Distribution

```python


thinkplot.PrePlot(3)

mus = [1.0, 2.0, 3.0]
sigmas = [0.5, 0.4, 0.3]
for mu, sigma in zip(mus, sigmas):
    xs, ps = thinkstats2.RenderNormalCdf(mu=mu, sigma=sigma, 
                                               low=-1.0, high=4.0)
    label = r'$\mu=%g$, $\sigma=%g$' % (mu, sigma)
    thinkplot.Plot(xs, ps, label=label)

thinkplot.Config(title='Normal CDF', xlabel='x', ylabel='CDF',
                 loc='upper left')
```

```python
# estimate parameters: trimming outliers yields a better fit
mu, var = thinkstats2.TrimmedMeanVar(fatal_billion, p=0.01)
print('Mean, Var', mu, var)
    
# plot the model
sigma = np.sqrt(var)
print('Sigma', sigma)
xs, ps = thinkstats2.RenderNormalCdf(mu, sigma, low=0, high=12.5)

thinkplot.Plot(xs, ps, label='model', color='0.6')

# plot the data
cdf = thinkstats2.Cdf(fatal_billion, label='data')

thinkplot.PrePlot(1)
thinkplot.Cdf(cdf) 
thinkplot.Config(title='fatal collisions per billion miles',
                 xlabel='fatal collisions',
                 ylabel='CDF')
```

```python
n = 1000
thinkplot.PrePlot(3) 

mus = [0, 1, 5]
sigmas = [1, 1, 2]

for mu, sigma in zip(mus, sigmas):
    sample = np.random.normal(mu, sigma, n)
    xs, ys = thinkstats2.NormalProbability(sample)
    label = '$\mu=%d$, $\sigma=%d$' % (mu, sigma)
    thinkplot.Plot(xs, ys, label=label)

thinkplot.Config(title='Normal probability plot',
                 xlabel='standard normal sample',
                 ylabel='sample values')
```

This plot suggests that the normal model describes the distributions fairly well with a few standard deviations from the mean, but not in the head.
