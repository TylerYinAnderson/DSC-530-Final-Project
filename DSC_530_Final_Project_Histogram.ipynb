```python
# Include a histogram of each of the 5 variables â€“ in your summary and analysis, 
# identify any outliers and explain the reasoning for them being outliers 
# and how you believe they should be handled (Chapter 2).

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('bad-drivers.csv')

fatal_billion = df['Number of drivers involved in fatal collisions per billion miles']
fatal_speed = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding']
fatal_alcoholimpared = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired']
fatal_notdistracted = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted']
fatal_notprevaccidents = df['Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents']



```

```python
plt.hist(fatal_billion)
plt.xlabel('Deaths Per Billion Miles')
plt.ylabel('Frequency')
plt.title('Number of drivers involved in fatal collisions per billion miles')
plt.show()
```

```python
plt.hist(fatal_speed)
plt.xlabel('% deaths by speeding')
plt.ylabel('Frequency')
plt.title('Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding')
plt.show()

# This shows a right leaning tail
```

```python
plt.hist(fatal_notdistracted)
plt.xlabel('% deaths from drivers who were not distracted')
plt.ylabel('Frequency')
plt.title('Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted')
plt.show()

# This shows a left leaning tail
```

```python
plt.hist(fatal_alcoholimpared)
plt.xlabel('% deaths from drivers who were alcohol impared')
plt.ylabel('Frequency')
plt.title('Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired')
plt.show()

# Tail is slightly to the right 
```

```python
plt.hist(fatal_notprevaccidents)
plt.xlabel('% deaths from drivers who had not been in an accident')
plt.ylabel('Frequency')
plt.title('Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents')
plt.show()
```
